```python
#This app generates a random number based upon the range selected by the user
#allow user to select a range for the number
    #user input for 2 numbers for the range
    #a button for calculating the random number
    #a button for clearing the data
    
#Updates from Activity 8:
    #every number chosen in a range is stored in a database
    #there are now scales for the range selection, rather than regular input fields

import tkinter as tk
from tkinter import *
from random import randint  
import sqlite3

#Creates a database to store the data in, once outputted
sql_connect = sqlite3.connect('database.db')
cur = sql_connect.cursor()

cur.execute('''CREATE TABLE IF NOT EXISTS random_data (
min_range INTEGER,
max_range INTEGER,
random_num INTEGER
)
''')
sql_connect.commit()

main = tk.Tk()

min_range_var = tk.IntVar() 
max_range_var = tk.IntVar()

label_welcome_msg = tk.Label(main, text="Please select 2 integers and a random number will be generated in that range.\n",
                 font=('helvetica',14,'bold'))

label_min_range = tk.Label(main, text="Minimum integer value for the range:",
                 font=('helvetica',14,'bold'))

min_range_scale = Scale(main, variable = min_range_var,  
           from_ = 0, to = 100,  
           orient = HORIZONTAL)  

label_max_range = tk.Label(main, text="Maximum integer value for the range:",
                 font=('helvetica',14,'bold'))

max_range_scale = Scale(main, variable = max_range_var,  
           from_ = 0, to = 100,  
           orient = HORIZONTAL) 

#This function's purpose is to clear the input portions
def clearData():
    min_range_scale.set(0)
    max_range_scale.set(0)

#This function produces a random number according to the set range
def getRandom():
    min_range = int(min_range_scale.get())
    max_range = int(max_range_scale.get())
    
    #If the minimum provided is greater than the maximum, then produce an error message
    if min_range > max_range:
        label_error_msg = tk.Label(main, text="An error occurred due to the minimum range being greater than the maximum range.\nPlease click the clear button to start over.",
                     font=('helvetica',14,'bold'))
        label_error_msg.pack()
    #If no error occurs, then produce a random number
    else:
        num = randint(min_range, max_range)
        label_response = tk.Label(main, text="The random number generated is: " + str(num),
                     font=('helvetica',14,'bold'))
        label_response.pack()
        
        #Insert the data into the database created earlier in the program
        cur.execute("INSERT INTO random_data (min_range,max_range,random_num) VALUES (?,?,?)",
                (min_range,max_range,num))
        sql_connect.commit()

#This function's purpose is to clear the database
def clearDB():
    cur.execute("DELETE FROM random_data")
    sql_connect.commit()
    
#This function shows the past data entered into the database
def showDataButton():
    cur.execute("SELECT * FROM random_data")
    collected_data = cur.fetchall()
    sql_connect.commit()
    
    #If the database is empty, then notify the user; otherwise show the data and the key about how the info is organized
    if collected_data == []:
        label_empty_data = tk.Label(main, text="The database is empty",
                     font=('helvetica',14,'bold'))
        label_empty_data.pack()
    else:
        label_data_info = tk.Label(main, text="The data is displayed in min, max, random number format.",
                         font=('helvetica',14,'bold'))
        label_data_info.pack()
        label_show_data = tk.Label(main, text=collected_data,
                         font=('helvetica',14,'bold'))
        label_show_data.pack()
        
#A button that clears the inputs
clear_inputs_button = tk.Button(main,
    text="Clear Scales",
    font=('calibre',14,'normal'),
    command = clearData    
)

#A button that submits the inputted data and runs the getRandom() function
submit_button = tk.Button(main,
    text="Submit",
    font=('calibre',14,'normal'),
    command = getRandom    
)

#A button that clears the database
clear_db_button = tk.Button(main,
    text="Clear Database",
    font=('calibre',14,'normal'),
    command = clearDB    
)

#A button that shows the database entries
show_data_button = tk.Button(main,
    text="Show Past Entries",
    font=('calibre',14,'normal'),
    command = showDataButton    
)

label_welcome_msg.pack()
label_min_range.pack()
min_range_scale.pack()
label_max_range.pack()
max_range_scale.pack()
clear_inputs_button.pack()
submit_button.pack()
clear_db_button.pack()
show_data_button.pack()
main.mainloop()
```
